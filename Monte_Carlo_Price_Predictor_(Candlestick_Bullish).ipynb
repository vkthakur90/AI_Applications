```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mplfinance as mpf
import datetime as dt
```

```python
mu = 0.10
sigma = 0.01
S0 = 100
year = 1
num_path = 50
```

```python
def plot_candle_sticks(mu, sigma, S0, year, num_path):
    e = np.random.normal(size = [365 * year + 1, num_path])
    t = np.arange(365 * year + 1)[..., np.newaxis] / 365

    delta_t = 1.0/365.0

    prev_S = S0 * np.ones_like(e[0, ...])

    data_list = [prev_S]

    for idx in range(365 * year + 1):
        next_S = prev_S * np.exp(mu * delta_t + sigma * np.sqrt(delta_t) * e[idx, ...])
        data_list += [next_S]
        prev_S = next_S

    S = np.stack(data_list, axis = 0)

    date_index = pd.DatetimeIndex([dt.date(2025, 1, 1) + dt.timedelta(delta) for delta in range(num_path)])

    candlestick_df = pd.DataFrame({
        'Open' : S[0, ...],
        'High' : np.max(S, axis = 0, keepdims = False),
        'Low'  : np.min(S, axis = 0, keepdims = False),
        'Close' : S[-1, ...]
    }, index = date_index)

    mpf.plot(
        candlestick_df,
        type="candle",  # Candlestick chart
        style="yahoo",  # Chart style
    )
```

```python
plot_candle_sticks(0.1, 0.0, S0, year, num_path)
```
